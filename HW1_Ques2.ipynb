```python
import numpy as np
import matplotlib.pyplot as plt

#importing Necessary libaries
```

```python
tInitial = int(input())
tFinal = int(input())
delT = float(input())

#Values are taken from users, but for this problem the values are, tInitial = 0, tFinal =10, delT = 0.1
```

```python
arraySize = int ((tFinal - tInitial)/delT)
arraySize

#The numbers of index we need to solve this problem.
#It will help us to determine how many computaion we require to solve this problem.
```

```python
x = np.zeros(int(arraySize))
y = np.zeros(int(arraySize))
theta = np.ones(int(arraySize))
v = np.ones(int(arraySize))
omega = np.zeros(arraySize)

x,y,theta, v, omega
#initializing x,y with 0. we need total 100 values of x,y,theta because ((10-0)/0.1)) = 100.
# theta is filled with 1 because the inital value of theta is 1
#As "v" has the value "1" in all cases, that's why it was initialized with value "1".
```

```python
# Conditions for Omega in different time cases.
#N.B: the all the timeslot has been multiplied by 10 for easier understanding.


for i in range(100):
  omega[i] = 0

  if i>=5 and i<=15:
    omega[i] = 3
  if i>=20 and i<=30:
    omega[i] = -3
  if i>=40 and i<=50:
   omega[i] = -3
  if i>=60 and i<=70:
    omega[i] = 3
  if i>=80 and i<=90:
   omega[i] = -3
```

```python
   print("New values of Omega:  ")
   omega
```

```python
# the differential equaions for the unicycle robot model using Eulerâ€™s method

for i in range(99):
  x[i+1] = x[i] + (0.1 * (np.cos(theta[i])) * v[i])
  y[i+1] = y[i] + (0.1 * (np.sin(theta[i])) * v[i])
  theta[i+1] = theta[i] + (delT * omega[i])

#N.B: in this case, we have iterate the loops 99 times because the initial value is already given. we  just need to find the next 99 values of x,y, theta.
```

```python
# Y vs T Graph
time_values = np.linspace(0, 10, len(y))

# Plotting the data
plt.plot(time_values, y, label='Y values')

# Add labels and legend
plt.xlabel('Time')
plt.ylabel('Y values')
plt.title(' Y vs T Graph ')
plt.legend()
plt.show()
```

```python
# Theta vs T Graph
time_values = np.linspace(0, 10, len(theta))

# Plotting the data
plt.plot(time_values, theta, label='Theta values')

# Add labels and legend
plt.xlabel('Time')
plt.ylabel('Theta values')
plt.title(' Theta vs T Graph ')
plt.legend()
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from matplotlib.animation import PillowWriter


# Create the figure and axes
fig, ax = plt.subplots()
point, = ax.plot([], [], 'cs', markersize=10)

# Add grid lines
ax.grid(True)

def init():
    ax.set_xlim(min(x), max(x))
    ax.set_ylim(min(y), max(y))
    return point,

def update(frame):
    x_frame = x[frame]
    y_frame = y[frame]
    point.set_data(x_frame, y_frame)
    return point,

# Set up the animation
animation = FuncAnimation(fig, update, frames=len(x), init_func=init, blit=True)

# Save the animation as a GIF
animation.save('custom_animation_with_grid.gif', writer=PillowWriter(fps=5))

from IPython.display import Image
Image(filename='custom_animation_with_grid.gif')
```
